```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('paper.mplstyle')

from pion_decay import dGamma_dE_rest,dGamma_dE_lab,E_N_rest_bounds,E_N_rest_min,E_N_rest_max,N_HNL
from pion_decay import pi0_width,m_pi0
```

```python
m_N_range = np.linspace(0.01,0.5,5)*m_pi0

for m in m_N_range:
    E_range = np.linspace(E_N_rest_min(m),E_N_rest_max(m),1000)
    dGamma_rest = []
    for E in E_range:
        dGamma_rest.append(dGamma_dE_rest(E,3e-7,m))
    print(sum(np.array(dGamma_rest)[np.array(dGamma_rest)>0])*(E_range[1]-E_range[0]))
    plt.plot(1e3*E_range,dGamma_rest,label=r'$m_\mathcal{N} = %2.1f$ MeV'%(m*1e3))
plt.semilogy()
plt.xlim(0,90)
plt.xlabel(r'$E_\mathcal{N}^{\rm Rest}$ [MeV]')
plt.ylabel(r'$\frac{1}{\Gamma} ~\frac{d \Gamma (\pi^0 \to \mathcal{N}\nu \gamma)}{d E_\mathcal{N}}$')
plt.title(r'$\pi^0$ Rest Frame')
plt.legend()
plt.show()

```

```python
for gamma in [1.00001,1.1,1.5,2,10]:
    beta = np.sqrt(1-gamma**2)
    for m in m_N_range:
        E_range = np.linspace(gamma*E_N_rest_min(m),gamma*E_N_rest_max(m),100)
        dGamma_lab = []
        for E in E_range:
            dGamma_lab.append(dGamma_dE_lab(E,3e-7,m,gamma))
        plt.plot(1e3*E_range,dGamma_lab,label=r'$m_\mathcal{N} = %2.1f$ MeV'%(m*1e3))
        print(sum(np.array(dGamma_lab)[np.array(dGamma_lab)>0])*(E_range[1]-E_range[0]))
    #plt.xlim(0,135)
    plt.semilogy()
    plt.xlabel(r'$E_\mathcal{N}^{\rm Lab}$ [MeV]')
    plt.ylabel(r'$\frac{1}{\Gamma} ~\frac{d \Gamma (\pi^0 \to \mathcal{N}\nu \gamma)}{d E_\mathcal{N}}$')
    plt.title('Lab Frame')
    plt.legend(title=r'$\gamma = %1.1f$'%gamma)
    plt.show()
```
