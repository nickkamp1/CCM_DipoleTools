```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('paper.mplstyle')

from pion_decay import dGamma_dE_rest,dGamma_dE_lab,E_N_rest_bounds,E_N_rest_min,E_N_rest_max
from pion_decay import pi0_width,m_pi0

def EN_max(m_N):
    return (m_N**2 + m_pi0**2) / (2*m_pi0)
```

```python
m_N_range = np.linspace(0.01,0.5,3)*m_pi0

for m in m_N_range:
    E_range = np.linspace(E_N_rest_min(m),E_N_rest_max(m),1000)
    dGamma_rest = []
    for E in E_range:
        dGamma_rest.append(dGamma_dE_rest(E,1e-6,m)/pi0_width)
    plt.plot(1e3*E_range,dGamma_rest,label=r'$m_\mathcal{N} = %2.1f$ MeV'%(m*1e3))
plt.xlim(0,135)
plt.xlabel(r'$E_\mathcal{N}^{\rm Rest}$ [MeV]')
plt.ylabel(r'$\frac{1}{\Gamma} ~\frac{d \Gamma (\pi^0 \to \mathcal{N}\nu \gamma)}{d E_\mathcal{N}}$')
plt.title(r'$\pi^0$ Rest Frame')
plt.legend()
plt.show()

```

```python
gamma = 10

for m in m_N_range:
    E_range = np.linspace(1.01*m,gamma*E_N_rest_max(m),100)
    dGamma_lab = []
    for E in E_range:
        dGamma_lab.append(dGamma_dE_lab(1e-6,m,E,gamma)/pi0_width)
    plt.plot(1e3*E_range,dGamma_lab,label=r'$m_\mathcal{N} = %2.1f$ MeV'%(m*1e3))
#plt.xlim(0,135)
plt.semilogy()
plt.xlabel(r'$E_\mathcal{N}^{\rm Lab}$ [MeV]')
plt.ylabel(r'$\frac{1}{\Gamma} ~\frac{d \Gamma (\pi^0 \to \mathcal{N}\nu \gamma)}{d E_\mathcal{N}}$')
plt.title('Lab Frame')
plt.legend()
plt.show()
```
