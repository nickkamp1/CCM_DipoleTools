```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('paper.mplstyle')

from pion_decay import dGamma_dE_rest,dGamma_dE_lab,pi0_width
```

```python
m_N_range = np.logspace(-2,-1,5)

for m in m_N_range:
    E_range = np.linspace(m,0.135,1000)
    dGamma_rest = []
    for E in E_range:
        dGamma_rest.append(dGamma_dE_rest(E,1e-6,m)/pi0_width)
    plt.plot(1e3*E_range,dGamma_rest,label=r'$m_\mathcal{N} = %2.1f$ MeV'%(m*1e3))
plt.ylim(0)
plt.xlabel(r'$E_\mathcal{N}$ [MeV]')
plt.ylabel(r'$\frac{1}{\Gamma} ~\frac{d \Gamma (\pi^0 \to \mathcal{N}\nu \gamma)}{d E_\mathcal{N}}$')
plt.title(r'$\pi^0$ Rest Frame')
plt.legend()
plt.show()

for m in m_N_range:
    E_range = np.linspace(m,0.135,1000)
    dGamma_lab = []
    for E in E_range:
        dGamma_lab.append(dGamma_dE_lab(1e-6,m,E,2)/pi0_width)
    plt.plot(1e3*E_range,dGamma_rest,label=r'$m_\mathcal{N} = %2.1f$ MeV'%(m*1e3))
plt.ylim(0)
plt.xlabel(r'$E_\mathcal{N}$ [MeV]')
plt.ylabel(r'$\frac{1}{\Gamma} ~\frac{d \Gamma (\pi^0 \to \mathcal{N}\nu \gamma)}{d E_\mathcal{N}}$')
plt.title(r'$\pi^0$ Rest Frame')
plt.legend()
plt.show()
```
